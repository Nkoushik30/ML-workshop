**-->importing required libraries.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# I choose youtubers dataset
ds1=pd.read_csv("/youtubers_df.csv")
pd.DataFrame(ds1)
ds2=ds1.drop(columns=["Categories"])
print(ds2)
print(ds2.info())
```



      *To read the data set provided by us will be read by the method
        read_csv()
      *Getting the basic information about the dataset will be given
        by the info() method.

**getting some samples and some sort of first 5 and last 5.**

```python
print(ds2.head())
print(ds2.tail())
print(ds2.shape)
print(ds2.columns)
ds2.sample(3)
ds2.dtypes
```



        * head(),tail() methods will give you the first 5 and last 5
          records
          to showoff how the dataset is present.
        * .shape and .column returns shape and column names present in
          the dataset.

**Bar graph between the number of subscribers and country.**

```python
subscribers_by_country = ds2.groupby("Country")["Subscribers"].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
subscribers_by_country.plot(kind="bar", edgecolor="black")
plt.title("Number of Subscribers by Country", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Subscribers", fontsize=12)
plt.xticks(rotation=90, ha="right")
plt.tight_layout()
plt.show()
```


      * The above shows us the relation between country and the number
        of subscribers on youtube for top 1000 youtuber's.

**lineplot between the subscribers and likes**

```python
subscribers=ds2['Subscribers']
likes=ds2['Likes']
plt.figure(figsize=(10, 6))
plt.plot(subscribers, likes, marker='o', linestyle='-', color='b')
plt.title('Subscribers vs. Likes', fontsize=16)
plt.xlabel('Subscribers', fontsize=14)
plt.ylabel('Likes', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

* The above line graph represents relation betwen subscriber and likes
  given by them.

**scatterplot between the subscribers and likes,comments,visits**

```python
subscribers=ds2['Subscribers']
visits=ds2['Visits']
likes=ds2['Likes']
comments=ds2['Comments']
plt.figure(figsize=(15, 15))
plt.subplot(3, 1, 1)
plt.scatter(subscribers, visits, color='blue', alpha=0.6)
plt.title('Subscribers vs. Visits', fontsize=16)
plt.xlabel('Subscribers', fontsize=14)
plt.ylabel('Visits', fontsize=14)
plt.subplot(3, 1, 2)
plt.scatter(subscribers, likes, color='green', alpha=0.6)
plt.title('Subscribers vs. Likes', fontsize=16)
plt.xlabel('Subscribers', fontsize=14)
plt.ylabel('Likes', fontsize=14)
plt.subplot(3, 1, 3)
plt.scatter(subscribers, comments, color='red', alpha=0.6)
plt.title('Subscribers vs. Comments', fontsize=16)
plt.xlabel('Subscribers', fontsize=14)
plt.ylabel('Comments', fontsize=14)
plt.tight_layout()
plt.show()
```

      * The above three scatter plots represents the data about how
        many subscribers are visiting the video and how many are liking,putting comments the video.
